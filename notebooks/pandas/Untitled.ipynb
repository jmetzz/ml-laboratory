```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        "item_group_code": ["COOLING", "COOLING", "COOLING", "SHAVERS", "SHAVERS", "TV", "TV"],
        "country_code": ["DE", "BR", "JP", "DE", "DE", "DE", "JP"],
        "period_seq": [2033, 2033, 2033, 2033, 2033, 2033, 2033],
        "fact": [1, 2, 3, 4, 5, 6, 7],
    }
)
df
```

```python
config_df = pd.DataFrame(
    {
        "item_group_code": ["COOLING", "COOLING", "COOLING", "TV"],
        "country_code": ["DE", "BR", "JP", "JP"],
        "feature": [100, 200, 300, 300],
    }
)
config_df
```

```python
df.merge(config_df, on=["item_group_code", "country_code"], how="inner")
```

```python
df1 = pd.DataFrame({"lkey": ["foo", "bar", "baz", "foo"], "value": [1, 2, 3, 5]})
df2 = pd.DataFrame({"rkey": ["foo", "bar", "valueA", "valueB"], "value": [1, 2, 7, 8]})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, left_on="lkey", right_on="rkey")
```
